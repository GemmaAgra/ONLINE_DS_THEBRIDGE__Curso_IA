## Juego del Ahorcado
Desarrolla el juego del ahorcado en python. En cada turno irá mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _``` en sus posiciones correspondientes. 

Si falla, se mostrará el número de errores que lleva y, si lleva x errores, entonces se mostrará "Fin del juego".

1. Inicializa el juego: declara una palabra a adivinar y muestra por pantalla la palabra oculta.



```python
_ _ _ _ _ _ _ _
```

2. Funcionamiento turnos: mediante un bucle genera el sistema de turnos donde el usuario introduce una letra y le va mostrando su progreso y sus vidas.

3. Desarrolla la finalización de la partida, tanto si gana como si pierde.

````python
def jugar_ahorcado(palabra_secreta, vidas):
   
    palabra_secreta = palabra_secreta.upper()
    progreso = ["_"] * len(palabra_secreta)  # guardo cómo va la palabra en ese momento
    letras_intentadas = set()  # set no deja elems repes 
    errores = 0  # para comparar con vidas

    print(" *** JUEGO DEL AHORCADO *** ")
    print("Palabra a adivinar: ", " ".join(progreso))   # método q une elems de una lista en 1 string

    while errores < vidas and "_" in progreso:
        letra = input("\nIntroduce una letra: ").upper()

        # VL
        if len(letra) != 1 or not letra.isalpha():    # comprueba si un texto contiene solo letras
            print("Introduce solo 1 letra.")
            continue

        if letra in letras_intentadas:
            print("Ya intentaste esa letra.")
            continue

        letras_intentadas.add(letra)

        # Comprobación
        if letra in palabra_secreta:
            print("¡Acierto!")
            # revela la letra en todas sus pos dentro de la palabra
            for i in range(len(palabra_secreta)): 
                if palabra_secreta[i] == letra:   # mira en cada pos si coincide la letra y si es asi la mete
                    progreso[i] = letra   # sustituyo letra por el "_"
        else:
            errores += 1
            print(f"Fallaste. Errores: {errores}/{vidas}")

        print("Progreso:", " ".join(progreso))

    # Final del juego
    if "_" not in progreso:
        print("\n¡Has ganado! 🎉 La palabra era: ", palabra_secreta)
    else:
        print("\nFin del juego 💀")
        print("La palabra era:", palabra_secreta)


# Ej.
jugar_ahorcado("hola", 3)
````

4. Encapsula el juego en una función llamada ```jugar_ahorcado()``` con la palabra a adivinar y el número de vidas para seleccionar la dificultad como argumentos de entrada.
