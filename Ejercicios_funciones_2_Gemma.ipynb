## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def num_a_dia(num):
    dias = {
        1: "Lunes",
        2: "Martes",
        3: "Miércoles",
        4: "Jueves",
        5: "Viernes",
        6: "Sábado",
        7: "Domingo"
    }
    return dias.get(num, "Número no válido")

print(num_a_dia(1))
print(num_a_dia(5))
print(num_a_dia(8))
````

En el ejercicio 8 de flujos de control, creábamos una pirámide invertida de números desde el 5. Crea una función que replique el comportamiento de la pirámide, y utiliza un único parámetro de entrada de la función para determinar el número de filas de la pirámide.

5 4 3 2 1 

4 3 2 1 

3 2 1

````python
def piramide_invertida(n):
    for i in range(n, 0, -1):          # controla las filas   Range va de n hasta 1
        for j in range(i, 0, -1):      # imprime los nºs de cada fila, el -1 va en decremento
            print(j, end=" ")
        print()  # salto de línea

piramide_invertida(5)
````

## Ejercicio 3
Escribe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero sea  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def comparar(a, b):
    if  a == b: return "Son iguales"
    elif a > b: return "1º > 2º"
    else:       return "2º > 1º"

print(comparar(5, 5))
print(comparar(7, 3))
print(comparar(2, 8))
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea una letra que contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayúscula, como minúscula

````python
def contar_letra(texto, letra):
    cont = 0

    for caracter in texto:
        if caracter.lower() == letra.lower():
            cont += 1
    return cont

print(contar_letra("Soy Gemma", "m"))
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string, siendo la clave la letra y el valor el conteo.

````python
def contar_letras(texto):
    DiccCont = {}

    for letra in texto:
        letra = letra.lower()  # para no diferenciar mayúsculas/minúsculas
        if letra in DiccCont:  DiccCont[letra] += 1   # si la clave letra ya existe en el diccionario cont
        else:                  DiccCont[letra] = 1
    return DiccCont

print(contar_letras("Soy Gemma"))
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

Tendrá como salida la lista

````python
def modificar_lista(lista, comando, elemento=None):
    if comando == "add":
        if elemento is not None:  
            lista.append(elemento)
        else: return "Di un elemento a añadir"  # si es None
    elif comando == "remove":
        #if elemento is None:  return "Di un elemento a eliminar"
        if elemento in lista:   # si existe
            lista.remove(elemento)
        else:  print("El elemento ", elemento , "no existe en la lista")
    else:
        return "Comando no válido"
    return lista

mi_lista = [1, 2, 3]

print(modificar_lista(mi_lista, "add", 4))      # añade 4
print(modificar_lista(mi_lista, "remove", 2))   # elimina 2
print(modificar_lista(mi_lista, "remove", 10))  # no hace nada
````

````python
def crear_frase(*palabras):
    return " ".join(palabras)  # une los elementos con un espacio

print(crear_frase("Hola", "soy", "Gemma", "con", "Python"))


def crear_frase(*palabras):
    frase = ""
    for palabra in palabras:
        frase += palabra + " "
    return frase.strip()  # elimina el último espacio

print(crear_frase("Hoy", "hace", "sol "))
````

Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

````python
def fibonacci(n):
    if   n == 1:  return 0
    elif n == 2:  return 1
    else:         return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(2))  
print(fibonacci(4))  
print(fibonacci(5))  
print(fibonacci(8))  



def fibonacci(n, memo={}):
    if n in memo:
        return memo[n]

    if n == 1: return 0
    if n == 2: return 1

    memo[n] = fibonacci(n-1, memo) + fibonacci(n-2, memo)
    return memo[n]


print("-----")
print(fibonacci(2))  
print(fibonacci(4))  
print(fibonacci(5))  
print(fibonacci(8))  

# Cada nº es la suma de los 2 anteriores:  f(n) = f(n-1) + f(n-2)
# No me sale el mismo Rdo q en el planteamiento
````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el área de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math

def area_cuadrado(lado):
    return lado * lado

def area_triangulo(base, altura):
    return (base * altura) / 2

def area_circulo(radio):
    return math.pi * radio ** 2


# 2 círculos de radio 10 + un triángulo (base 3, altura 7)
rdo1 = (2 * area_circulo(10) + area_triangulo(3, 7))
print("Rdo 1: ", rdo1)


# Cuadrado de lado 10 + 3 círculos + 5 triángulos
rdo2 = (
    area_cuadrado(10) +
    area_circulo(4) +
    2 * area_circulo(6) +
    5 * area_triangulo(2, 4)
)
print("Rdo 2: ", rdo2)
````
