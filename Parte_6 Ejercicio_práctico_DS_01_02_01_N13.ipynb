`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un",
          "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

# elementos a buscar
busquedas = ["un", "es", "binario"]

for duplicado in busquedas:
    indices = []

    for i, valor in enumerate(ejer_3):
        if valor.lower() == duplicado:
            indices.append(i)

    print(f"Índices de '{duplicado}':", indices)
`````

## Ejercicio 9
1. Escribe un programa que dada una variable titulo busque los libros que hay en la librería con ese título. Si no se encuentra debe devolver el mensaje `Ese no lo tengo, ¿mola?`
2. Pruena que el programa con uno de tus libros y con otro que no contenga tu librería.

`````python
# Función de búsqueda por título
def buscar_libro(titulo_buscado, libreria):
    encontrado = False

    for libro in libreria:
        if libro["titulo"].lower() == titulo_buscado.lower():
            print("Libro encontrado:", libro)
            encontrado = True

    if not encontrado:   print("Ese no lo tengo")


buscar_libro("1984", libreria)           # libro que sí está en la librería
buscar_libro("Harry Potter", libreria)   # libro que NO está en la librería
`````

Convierte el programa del ejercicio 9 en una función al que se le pasen como argumentos la librería, posicional, y el título, argumento keyword con valor "ninguno", por defecto.

`````python
def buscar_libro(libreria, titulo="ninguno"):
    encontrado = False

    for libro in libreria:
        if libro["titulo"].lower() == titulo.lower():
            print("Libro encontrado:", libro)
            encontrado = True

    if not encontrado:  print("Ese no lo tengo")

# buscar un libro existente
buscar_libro(libreria, "1984")

# buscar un libro inexistente
buscar_libro(libreria, "Harry Potter")

# usar valor por defecto del keyword
buscar_libro(libreria)
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def dni_valido(dni):
    palabra = "TRWAGMYFPDXBNJZSQVHLCKE"

    # Separar nº y letra
    num = dni[:-1]
    letra = dni[-1].upper()

    # 1. Comprobar que el nº tenga entre 7 y 8 dígitos y sea numérico
    if not (num.isdigit() and 7 <= len(num) <= 8):
        return False

    # 2. Calcular letra correcta
    num_int = int(num)
    letra_correcta = palabra[num_int % 23]   # Calcula la letra correcta con nº % 23

    # 3. Comparar letras
    return letra == letra_correcta


# Ej.
dni = input("Introduce tu DNI (ej: 12345678Z): ")

if dni_valido(dni):
    print("DNI válido")
else:
    print("DNI no válido")
`````
