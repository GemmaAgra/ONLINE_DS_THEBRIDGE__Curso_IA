## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
Lista1=["Hola", "amigo"]   
Lista2=["Q", "tal"]
ListaRdo=[]

for i in Lista1:
    for ii in Lista2:
        ListaRdo.append(i + " " + ii)

print(ListaRdo)

# abreviado
rdo = [f"{a} {b}" for a in Lista1 
                    for b in Lista2]

print(rdo)
````

## Ejercicio 4
Dada la siguiente lista, encuentra por el índice el valor 45 y sustitúyelo por el 0
> ```ejer_4 = [20, 47, 19, 29, 45, 67, 78, 90]```

````python

Lista =[20,47,19,29,45,67,78,90]

Lista = [0 if x == 45 else x for x in Lista]
print(Lista)

# otra forma
for i in range(len(Lista)):
    if Lista[i] == 45:
        Lista[i] = 0

print(Lista)

Lista =[20,47,19,29,45,67,78,90]

#otra forma
Lista[Lista.index(45)] = 0
print(Lista)
````

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

TIP: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
lista = [3,20,3,47,19,3,29,45,67,78,90,3,3]
lista = [x for x in lista 
               if x != 3]
print(lista)

# otra forma
lista = [3,20,3,47,19,3,29,45,67,78,90,3,3]
while 3 in lista:
    lista.remove(3)
print(lista)
````

## Ejercicio 11
Convierte la siguiente lista en un diccionario
> ```ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
diccionario = {}

for clave, valor in ejer_11:
    if clave not in diccionario:
        diccionario[clave] = []
    diccionario[clave].append(valor)

print(diccionario)
````

## Ejercicio 15
Multiplica todos los valores del diccionario
> ```ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}```

````python
ejer_15 = {1: 11, 2: 22, 3: 33, 4: 44, 5: 55}

producto = 1

for valor in ejer_15.values():
    producto *= valor

print(producto)
````

## Ejercicio 16
1. Crea un set de tres elementos
2. Añade un cuarto
3. Elimina el último elemento añadido
