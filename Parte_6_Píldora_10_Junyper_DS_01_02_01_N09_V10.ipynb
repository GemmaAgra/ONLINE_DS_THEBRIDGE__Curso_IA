##  Argumentos clave-valor

### Contenidos

Para terminar con las diferentes posibilidades que ofrecen los argumentos de las funciones, vamos a tratar los argumentos clave-valor.  

Es decir, tenemos también la opción de introducir un diccionario como argumentos, de esta forma, aunque el numero de argumentos sea variable, tendremos un inidicador, la clave, y el valor de cada clave.

Se implementa con `**`

```python
# mi Ej.

def mostrar_datos(**datos):
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")

mostrar_datos(nombre="Gemma", edad=25, ciudad="Vitoria")
```

### Combinar `*args` con `**kwargs`
[al indice](#Contenidos)  
No hay ningun problema en tener un numero variable de argumentos y también argumentos clave-valor, todo ello en la misma función.

```python
def ej(*args, **kwargs):
    print("Args posicionales (*args):")
    for arg in args:
        print(arg)

    print("\nArgumentos clave-valor (**kwargs):")
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")

ej(10, 20, 30, nombre="Gemma", edad=25, ciudad="Vitoria")
```

### Argumentos keyword
[al indice](#Contenidos)  
Existe otro tipo de argumentos que son los *keyword*. Se caracterizan porque llevan un valor por defecto, y por tanto, si no usamos dicho argumento en la llamada, dentro de la función tomará el valor que hayamos dejado por defecto.

Ten en cuenta que estos argumentos se colocan **al final**

```python
#def funcion(obligatorio, opcional="valor", opcional2=10):
def saludar(nombre, saludo="Hola"):
    print(f"{saludo}, {nombre}")

saludar("Gemma")   # valor por defecto  
saludar("Ana", saludo="Buenos días")    # Cambiando el argumento keyword
```

```python
# Arg con valor por defecto antes de uno obligatorio
# Un arg obligatorio (nombre) no puede ir después de uno con valor por defecto
def saludar(saludo="Hola", nombre):    
    print(saludo, nombre)

# Pasar argus por posición y keyword mal combinados
# 1º van los posicionales, luego los keyword.
def saludar(nombre, saludo="Hola"):
    print(saludo, nombre)
saludar(saludo="Buenos días", "Gemma")

# Usar un keyword que no existe definido en la función
def saludar(nombre, saludo="Hola"):
    print(saludo, nombre)
saludar(nombre="Ana", mensaje="Buenos días")

# Olvidar valor obligatorio. nombre sin valor por defecto, así que es obligatorio.
def saludar(nombre, saludo="Hola"):
    print(saludo, nombre)
saludar()

# Default antes de obligatorio → error
# Posicional después de keyword → error
# Keyword inexistente → error
# Faltan obligatorios → error
```
